```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd 
import numpy as np
from math import *
import matplotlib.pyplot as plt
data = pd.read_csv('Fast Gait No Arms 01.csv')
data = data.apply(pd.to_numeric, errors='coerce')
labels =  ['LFHD','RFHD','LBHD','RBHD','C7','T10','CLAV','STRN','RBAK','LSHO','LUPA','LELB','LFRM','LWRA','LWRB','LFIN','RSHO','RUPA','RELB','RFRM',
'RWRA','RWRB','RFIN','LASI','RASI','LPSI','RPSI','LTHI','LKNE','LTIB','LANK','LHEE','LTOE','RTHI','RKNE','RTIB','RANK','RHEE','RTOE']
data
```

```python
#Head
LFHD =  data.iloc[:,2:5]
RFHD =  data.iloc[:,5:8]
LBHD =  data.iloc[:,8:11]
RBHD =  data.iloc[:,11:14]

#Front and Back
C7 =  data.iloc[:,14:17]
T10 =  data.iloc[:,17:20]
CLAY =  data.iloc[:,20:23]
STRN =  data.iloc[:,23:26]
RBAK =  data.iloc[:,26:29]

#Left Arm
LSHO =  data.iloc[:,29:32]
LUPA =  data.iloc[:,32:35]
LELB =  data.iloc[:,35:38]
LFRM =  data.iloc[:,38:41]
LWRA =  data.iloc[:,41:44]
LWRB =  data.iloc[:,44:47]
LFIN =  data.iloc[:,47:50]

#Right Arm
RSHO =  data.iloc[:,50:53]
RUPA =  data.iloc[:,53:56]
RELB =  data.iloc[:,56:59]
RFRM =  data.iloc[:,59:62]
RWRA =  data.iloc[:,62:65]
RWRB =  data.iloc[:,65:68]
RFIN =  data.iloc[:,68:71]

#Sides
LASI =  data.iloc[:,71:74]
RASI =  data.iloc[:,74:77]
LPSI =  data.iloc[:,77:80]
RPSI =  data.iloc[:,80:83]

#Left Leg
LTHI =  data.iloc[:,83:86]
LKNE =  data.iloc[:,86:89]
LTIB =  data.iloc[:,89:92]
LANK =  data.iloc[:,92:95]
LHEE =  data.iloc[:,95:98]
LTOE =  data.iloc[:,98:101]

#Right Leg
RTHI =  data.iloc[:,101:104]
RKNE =  data.iloc[:,104:107]
RTIB =  data.iloc[:,107:110]
RANK =  data.iloc[:,110:113]
RHEE =  data.iloc[:,113:116]
RTOE =  data.iloc[:,116:119]
```

```python
#Calulating Linear Velocity
#Formula = (Xf - Xi)/ Δt 
#t = 0.01

#Subtracting next row by current row 
diff = data.diff()
#Removing First 2 Columns and Rows
diff = diff.loc[:, ~diff.columns.isin(['Frame','Sub Frame'])]
diff = diff.tail(diff.shape[0] -2)

#Squaring every value -> Adding every 3 colums -> dividing by the time (0.01) to get linear velocity 
diff = (diff / 0.01) ** 2
linear_velocity = (diff.groupby((np.arange(len(diff.columns)) // 3)+1, axis=1).sum())
linear_velocity = (linear_velocity) ** (1/2)
linear_velocity = linear_velocity.drop(columns = linear_velocity.columns[-1])
linear_velocity.columns = labels
linear_velocity
```

```python
plt.figure
plt.rcParams['figure.figsize'] = (8,4)
y1 = linear_velocity['LTHI']
y2 = linear_velocity['LKNE']
y3 = linear_velocity['LTIB']
y4 = linear_velocity['LANK']

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.plot(y4)

plt.title('Linear Velocity of Left Leg')
plt.xlabel('Frames(1 = 0.01 sec)')
plt.ylabel('Velocity (mm/s)')
plt.legend(['LTHI','LKNE','LTIB','LANK'])
```

```python
plt.figure
plt.rcParams['figure.figsize'] = (8,4)
y1 = linear_velocity['RTHI']
y2 = linear_velocity['RKNE']
y3 = linear_velocity['RTIB']
y4 = linear_velocity['RANK']

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.plot(y4)
plt.grid()
plt.title('Linear Velocity of Right Leg')
plt.xlabel('Frames(1 = 0.01 sec)')
plt.ylabel('Velocity (mm/s)')
plt.legend(['RTHI','RKNE','RTIB','RANK'])
```

```python
#Calculating Linear Acceleration
#Formula = (Vf - Vi)/ Δt 
#t = 0.01
linear_acceleration = (linear_velocity.diff()) / 0.01
linear_acceleration = linear_acceleration.tail(linear_acceleration.shape[0] - 1)
linear_acceleration.columns = labels
linear_acceleration
```

```python
plt.figure
plt.rcParams['figure.figsize'] = (12,4)
y1 = linear_acceleration['LTHI']
y2 = linear_acceleration['LKNE']
y3 = linear_acceleration['LTIB']
y4 = linear_acceleration['LANK']

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)
plt.plot(y4)

plt.title('Linear Acceleration of Left Leg')
plt.xlabel('Frames(1 = 0.01 sec)')
plt.ylabel(r'$Acceleration (mm/s^2)$')
plt.legend(['LTHI','LKNE','LTIB','LANK'])
```

```python
#Calculating Angular Velocity between LKNE and LTHI
temp = ['0','1','2']
LKNE.columns = temp
LTHI.columns = temp
W = LTHI - LKNE
#Numerator of theta fraction
cos_theta_numerator = pd.DataFrame(columns = temp)
for i in range(0,723):
    temp1 = pd.DataFrame(W.iloc[i,:] * W.iloc[i+1,:])
    temp1 = temp1.transpose()
    cos_theta_numerator = cos_theta_numerator.append(temp1)
cos_theta_numerator = (cos_theta_numerator.groupby((np.arange(len(cos_theta_numerator.columns)) // 3)+1, axis=1).sum())
#Denominator of theta fraction
temp2 = W ** 2
temp2 = (temp2.groupby((np.arange(len(temp2.columns)) // 3)+1, axis=1).sum())
cos_theta_denominator = pd.DataFrame(columns = [1])
for i in range(0,723):
    denominator = pd.DataFrame((temp2.iloc[i,:] ** (1/2)) * (temp2.iloc[i+1,:] ** (1/2)))
    denominator = denominator.transpose()
    cos_theta_denominator = cos_theta_denominator.append(denominator)
#Finding theta in radians
cos_theta = cos_theta_numerator/cos_theta_denominator
#Converting theta to degrees
cos_theta = cos_theta.astype(float)
cos_theta = np.arctan(cos_theta)
cos_theta = np.degrees(cos_theta)
W_LTHI = cos_theta.diff() / 0.01 
W_LTHI = W_LTHI.tail(W_LTHI.shape[0] - 1)
W_LTHI.columns = ['W_LTHI']
W_LTHI.reset_index(drop=True, inplace=True)
```
